```python
#import các thư viện cần sử dụng
import numpy as np
import pandas as pd
import csv
```

## 1.

```python
#Đọc File
housing_df = pd.read_csv("D:\DA07\Bai3.1\VN_housing_dataset.csv", encoding="utf-8")
housing_df
```

## 2.



```python
# Kiểm tra các t tin bất thường
pd.isna(housing_df)

#loại bỏ nếu thấy bất hợp lý
housing_df.dropna(inplace = True)
housing_df[0:10:]
```

## 3.

```python
# Thay đơn vị "tỷ " sang "triệu"
housing_df.replace("tỷ/m²","triệu/m²")
housing_df
```

```python
#lấy t hông tin từ cột Ngày Tháng
import datetime as dt

housing_df['Ngày'] = pd.to_datetime(housing_df.Ngày, format='%Y-%m-%d')
#format lại dinh dạng của thành Ngày/Tháng/Năm
housing_df['Ngày'] = housing_df['Ngày'].dt.strftime('%d-%m-%Y')
```
